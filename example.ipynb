# Examples

Here are some examples. You will need to install Jupyter Notebook to get this notebook running again.

```python
import pandas as pd
from epldata.api import EPLAPI
```

```python
epl = EPLAPI()
season_id = epl.get_season_id("2024/25")
season_id
```

```python
stat_types = epl.list_available_stat_types()
stat_types
```

```python
goals = epl.get_player_stats("goals", season_id, output="df")
goals
```

```python
all_players = epl.list_all_players(season_id,output='df')
all_players
```

```python
search_result = epl.search_player_by_name("Mohamed Salah", season_id)
search_result
```

```python
mo_id = epl.get_player_id("Mohamed Salah",season_id)
mo_id
```

```python
player_id = search_result[0]['ID']
player_id
```

```python
stats_df = epl.get_player_info(player_id, season_id)
stats_df
```

```python
entity = stats_df['entity']
stats_list = stats_df['stats']

# 将 stats 转换成字典方便查找
stats_dict = {stat['name']: stat['value'] for stat in stats_list}

# 提取字段
player_summary = {
    "name": entity['name']['display'],
    "position": entity['info']['position'],
    "shirtNum": entity['info'].get('shirtNum'),
    "age": entity.get('age'),
    "appearances": stats_dict.get('appearances'),
    "goals": stats_dict.get('goals'),
    "goal_assist": stats_dict.get('goal_assist')
}

df_player = pd.DataFrame([player_summary])
df_player
```

```python
club_id = epl.get_club_id(season_id,"df")
club_id
```

```python
comp = epl.club_incompseason(season_id)
comp
```

```python
club_playedgames = epl.club_playedgames(season_id,10)
club_playedgames
```

```python
club_information = epl.club_information(10)
club_information
```
